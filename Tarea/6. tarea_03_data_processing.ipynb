**P4.** Rellena los valores nulos de la columna `Age` con la **mediana** de esa columna.  
Verifica que ya no haya nulos en `Age`.

```python
# Tu respuesta aquí
rellenar = df["Age"].fillna(df["Age"].median())
rellenar.isnull().sum()
```

## Parte 3 — Nuevas columnas y agrupaciones

**P5.** Crea una nueva columna llamada `FamilySize` que sea la suma de `SibSp` + `Parch` + 1 (el pasajero mismo).

```python
# Tu respuesta aquí
familySize = df["SibSp"] + df["Parch"] + 1
familySize
```

**P6.** Usando `groupby`, calcula la **tasa de supervivencia promedio** (columna `Survived`) por clase (`Pclass`).  
¿Qué clase tuvo mayor probabilidad de sobrevivir?

```python
# Tu respuesta aquí
tasa_supervivencia = df.groupby("Pclass")["Survived"].mean()
print(f"la clase con mayor supervivencia fue: {tasa_supervivencia}")
```

**Respuesta:** La clase con mayor supervivencia fue: _____

**P7.** Calcula la edad promedio y la tarifa (`Fare`) promedio, agrupando por `Sex` y `Pclass`.  
Usa `.groupby()` con `.agg()`.

```python
# Tu respuesta aquí
edad_promedio = df.groupby(['Sex', 'Pclass']).agg({'Age': 'mean','Fare': 'mean'})
print(edad_promedio)
```
