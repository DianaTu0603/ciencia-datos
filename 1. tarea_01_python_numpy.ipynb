**P7.** Dado el arreglo `a = np.arange(1, 13)`, usa `reshape` para convertirlo en una matriz de 3 × 4.  
Luego imprime el valor máximo, el valor mínimo y la posición (índice) del máximo.

```python
a = np.arange(1, 13)

# Tu respuesta aquí
matriz_3x4 = a.reshape(3, 4)    
print(f"Matriz 3x4:\n{matriz_3x4}")
print(f"Valor máximo: {matriz_3x4.max()}")
print(f"Valor mínimo: {matriz_3x4.min()}")
print(f"La posición indice del máximo:{matriz_3x4.argmax()}")
```

**P8.** Crea dos arreglos de 5 elementos con `np.random.randint` (valores entre 1 y 20).  
Súmalos, réstalos y multiplícalos elemento por elemento. Imprime cada resultado.

```python
# Tu respuesta aquí
arreglo1 = np.random.randint(1, 10, size=5)
print(f"Arreglo 1: {arreglo1}")
arreglo2 = np.random.randint(1, 10, size=5)
print(f"Arreglo 2: {arreglo2}")
suma = arreglo1 + arreglo2
print(f"Suma elemento por elemento: {suma}")
resta = arreglo1 - arreglo2
print(f"Resta elemento por elemento: {resta}")
multiplicacion = arreglo1 * arreglo2
print(f"Multiplicación elemento por elemento: {multiplicacion}")  
```

**P9.** Dado el arreglo `notas = np.array([8.5, 6.0, 9.2, 4.3, 7.8, 5.5, 10.0, 3.1])`,  
usa **indexación booleana** para obtener solo las notas mayores o iguales a 6. Imprime el resultado.

```python
notas = np.array([8.5, 6.0, 9.2, 4.3, 7.8, 5.5, 10.0, 3.1])

# Tu respuesta aquí
notas_aprobadas = notas[notas >= 6.0]
print(f"Notas aprobadas: {notas_aprobadas}")
```
